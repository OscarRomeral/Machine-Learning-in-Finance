### Question 1 - Importing the data

```python
#For Question 1, you can import the following packages:
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import shapiro

#1.a) Import the dataset Housing.csv into Python as a pandas DataFrame and call it Housing.
#To determine whether a variable is categorical or numerical, please refer to the file description.txt.
#Remember that the first column of the csv file is an index column and should not be considered as an explanatory variable.
Housing = pd.read_csv("./Housing.csv").drop(columns='Id')
Housing.head()
```

```python
import seaborn as sns

#1.b) Determine graphically whether the target variable SalePrice exhibits a Gaussian distribution. 
salePrice = Housing['SalePrice']

# Plot the histogram of SalePrice
plt.figure(figsize=(10,6))
sns.histplot(salePrice, kde=True, bins=30)
plt.title('Histogram of SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('Frequency')
plt.show()

# Create a Q-Q plot to check if the data is normally distributed
plt.figure(figsize=(6,6))
stats.probplot(salePrice, dist="norm", plot=plt)
plt.title('Q-Q Plot of SalePrice')
plt.show()
```

```python
# As we can see in both plots, the distribution of SalePrice doesn't fit a Gaussian
# We will try different transformations and see which ones exhibit better results regarding gaussianity


# Apply a log transformation if the data is skewed
logSalePrice = np.log(salePrice)

# 1. Plot the transformed SalePrice to check the distribution
plt.figure(figsize=(10,6))
sns.histplot(logSalePrice, kde=True, bins=30)
plt.title('Histogram of Log(SalePrice)')
plt.xlabel('Log(SalePrice)')
plt.ylabel('Frequency')
plt.show()

# 2. Create a Q-Q plot for the log-transformed data
plt.figure(figsize=(6,6))
stats.probplot(logSalePrice, dist="norm", plot=plt)
plt.title('Q-Q Plot of Log(SalePrice)')
plt.show()

# 3. Compute shapiro test
print(shapiro(logSalePrice))

# Apply a box-cox if the data is skewed
box_cox_salePrice, lambda_val = stats.boxcox(salePrice)

# 1. Plot the transformed SalePrice to check the distribution
plt.figure(figsize=(10,6))
sns.histplot(box_cox_salePrice, kde=True, bins=30)
plt.title('Histogram of box-cox SalePrice')
plt.xlabel('box-cox SalePrice')
plt.ylabel('Frequency')
plt.show()

# 2. Create a Q-Q plot for the box-cox transformed data
plt.figure(figsize=(6,6))
stats.probplot(box_cox_salePrice, dist="norm", plot=plt)
plt.title('Q-Q Plot of box-cox SalePrice')
plt.show()

# 3. Compute shapiro test
print(shapiro(box_cox_salePrice))
```

```python
# Both transformations improve the normality of the data. However, the low p-values from the Shapiro-Wilk tests indicate that the data still does not follow a perfect normal distribution.
# This is expected, as real-world price data often deviates from normality. Despite this, we can proceed with the Box-Cox transformation, as it offers a slightly better approximation of a Gaussian distribution.
```

```python
#We create a new column in the df with the transformed prices
Housing['SalePriceTrans'] = box_cox_salePrice
```

### Question 2 - Linear Regression on Housing2

```python
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm

#2.a) Split the Housing2 data into a training set (X,Y)_train and a test set (X, Y)_test. 
#Randomly assigned 70% of the observations to the training set and the remaining 30% to the test set.
X = Housing2.drop(['SalePrice','SalePriceTrans'], axis=1)  # Features (excluding the target variable)
Y = Housing2['SalePriceTrans']  # Target variable

# 2. Split the data into training (70%) and test (30%) sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
#2.b) Fit a linear regression model on the training dataset (X, Y)_train using the sklearn package.
#Output a table with the name of each feature and the associated regression coefficient. 

# Fit the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)

# Regression coefficients
coefficients = pd.DataFrame(model.coef_, X_train.columns, columns=['Coefficient'])
print(coefficients)
# Predicted values for training and test datasets
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)

# MSE and R2 for in-sample and out-of-sample
mse_train = mean_squared_error(Y_train, Y_train_pred)
mse_test = mean_squared_error(Y_test, Y_test_pred)

r2_train = r2_score(Y_train, Y_train_pred)
r2_test = r2_score(Y_test, Y_test_pred)

# Results presentation
results = {
    'Metric': ['Mean Squared Error', 'R^2'],
    'In-Sample': [mse_train, r2_train],
    'Out-of-Sample': [mse_test, r2_test]
}

results_df = pd.DataFrame(results)
print(results_df)
```

```python
#2.c) (i) Compute the estimated coefficients for each explanatory variable using the numpy package. Remember that the design matrix A needs to include a column of 1's.
A = np.array(X_train)
# Adddition of column of ones for the intercept term
A_with_intercept = np.hstack((np.ones((A.shape[0], 1)), A))


# Target variable vector y (reshape to have a 2D column vector)
y = np.array(Y_train).reshape(-1, 1)

# Estimated coefficients Î²Ì‚
beta_hat = np.linalg.inv(A_with_intercept.T @ A_with_intercept) @ A_with_intercept.T @ y

# Save resurt as a DataFrame
beta_hat_df = pd.DataFrame(beta_hat, columns=['Coefficient'], index=['Intercept'] + list(X_train.columns))

print("Estimated Coefficients (Î²Ì‚):")
print(beta_hat)
print(beta_hat_df)
```

```python
#	  (ii) Compute the standard error of each estimated coefficient using the numpy package.
# Predicted values Å·
y_pred = A_with_intercept @ beta_hat

# Residuals
residuals = y - y_pred

m = A.shape[0]  # Number of observations
d = A.shape[1]  # Number of explanatory variables

residuals_sum = 0
for i in range(m):
    residuals_sum += residuals[i]
    
s_squared = residuals_sum / (m - (d + 1))

# 6. Calculate (A^TA)^-1 and extract diagonal
A_T_A_inv = np.linalg.inv(A_with_intercept.T @ A_with_intercept)
aj = np.diag(A_T_A_inv)

# 7. Compute standard errors for each coefficient
standard_errors = np.sqrt(s_squared * aj)

# Prepare standard errors for output
standard_errors_df = pd.DataFrame(standard_errors, columns=['Standard Error'], index=['Intercept'] + list(X_train.columns))

print("Standard Errors of Estimated Coefficients (SE(Î²Ì‚)): ")
print(standard_errors_df)


#     (iii) Compute the (in-sample) MSE and R^2 using the numpy package.

#     (iv) Do the results change using pseudoinversion instead of standard matrix inversion in (i) and (ii)?

#     (v) Compare your results (i), (ii), (iii) (with and without pseudoinversion) to the output of the sm.OLS function. 

#2.d) Suggest instead a second-order polynomial regression model (including all quadratic and mixed terms) to predict the (transformed) SalePrice.

#Does this improve the linear regression model 2.b)? 

#What happens with higher-order polynomial regression models?
```
